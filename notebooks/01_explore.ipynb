# PlantVision — Dataset Exploration

Quick EDA on PhenoBench before training.


```python
import sys; sys.path.insert(0, '../src')
import yaml
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from plantvision.ingest import load_config, verify_split, dataset_stats
```

```python
cfg = load_config('../configs/default.yaml')
train_stats = verify_split(Path(cfg['data']['train']), 'train')
val_stats   = verify_split(Path(cfg['data']['val']),   'val')
print(train_stats, val_stats)
```

```python
pixel_dist = dataset_stats(cfg)
labels = {0: 'background', 1: 'crop', 2: 'weed'}
plt.bar([labels.get(k, str(k)) for k in pixel_dist], pixel_dist.values())
plt.ylabel('Pixel share (%)')
plt.title('PhenoBench pixel class distribution (50-image sample)')
plt.tight_layout(); plt.show()
```

```python
# Visualise a random sample
import random
train_dir = Path(cfg['data']['train'])
samples = random.sample(sorted((train_dir / 'images').glob('*.png')), 4)

fig, axes = plt.subplots(4, 2, figsize=(10, 16))
for ax_row, img_path in zip(axes, samples):
    mask_path = train_dir / 'semantics' / img_path.name
    ax_row[0].imshow(Image.open(img_path)); ax_row[0].set_title('Image'); ax_row[0].axis('off')
    ax_row[1].imshow(Image.open(mask_path), vmin=0, vmax=2, cmap='Set1'); ax_row[1].set_title('Mask'); ax_row[1].axis('off')
plt.tight_layout(); plt.show()
```
